```python
import pandas as pd
import matplotlib.pyplot as plt
sn = pd.read_csv('social_network_edges.csv')
plt.plot(sn['source'], sn['target'])
plt.title('Social Network')
plt.xlabel('Source')
plt.ylabel('Target')
plt.show()

iris = pd.read_csv('iris_dataset.csv')
categories = iris['sepal length (cm)'].value_counts()
plt.bar(categories.index, categories.values)
plt.title('Iris Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Counts')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('social_network_edges.csv')
sns.scatterplot(x='source', y='target', data=data, hue='source', style='target')
plt.title('Social Network Scatter Plot')
plt.show()


sns.pairplot(iris, hue='petal length (cm)', diag_kind='kde')
plt.show()
```
