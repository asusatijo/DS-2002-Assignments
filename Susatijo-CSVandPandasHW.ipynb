```python
#1
import pandas as pd
df = pd.read_csv('/content/acc_players-2324F.csv')
df.columns = df.iloc[0]
df = df[1:]
df.reset_index(drop=True, inplace=True)
df.head()
```

```python
#2
total_points = df['PTS'].sum()
print(f"Total points scored by all players: {total_points}")
print("\n")
player_with_most_minutes = df.loc[df['MP'].idxmax()]
print(f"The player with the most minutes played is: {player_with_most_minutes['Player']}")
print("\n")
top_5_rebounders = df.sort_values('TRB', ascending=False).head()
print("Top 5 players in terms of total rebounds:")
print(top_5_rebounders[['Player', 'TRB']])
```

```python
#3
players_with_500_minutes = df[df['MP'].astype(int) > 500]
players_with_500_minutes.head()
```

```python
player_with_highest_assists = players_with_500_minutes.loc[players_with_500_minutes['AST'].idxmax()]
print(player_with_highest_assists['Player'])
```

```python
top_3_assist_leaders = players_with_500_minutes.sort_values('AST', ascending=False).head(3)
print(top_3_assist_leaders[['Player', 'AST']])
```

```python
top_3_shot_blockers = players_with_500_minutes.sort_values('BLK', ascending=False).head(3)
print(top_3_shot_blockers[['Player', 'BLK']])
```

```python
#4
school_total_points = df.groupby('School')['PTS'].sum()
print(school_total_points)
```

```python
school_total_assists = df.groupby('School')['AST'].sum()
print(school_total_assists)
```

```python
school_total_points.sort_values(ascending=False).head(3)
```

```python
import matplotlib.pyplot as plt

top_5_scorers = df.sort_values('PTS', ascending=False).head(5)

plt.figure(figsize=(10, 6))
plt.bar(top_5_scorers['Player'], top_5_scorers['PTS'])
plt.xlabel('Player')
plt.ylabel('Total Points Scored')
plt.title('Top 5 Players by Total Points Scored')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

df['FG%'] = pd.to_numeric(df['FG%'], errors='coerce')
df['WS'] = pd.to_numeric(df['WS'], errors='coerce')

df_cleaned = df.dropna(subset=['FG%', 'WS'])

correlation = np.corrcoef(df_cleaned['FG%'], df_cleaned['WS'])[0, 1]

plt.figure(figsize=(8, 6))
plt.scatter(df_cleaned['FG%'], df_cleaned['WS'])
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')
plt.title('Correlation between FG% and WS')
plt.grid(True)
plt.show()

print(f"Correlation coefficient between FG% and WS: {correlation:.2f}")
```
